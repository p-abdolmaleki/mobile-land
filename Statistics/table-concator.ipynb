```python
df = brand.merge(phone , left_on='id', right_on='brand_id')
df = df.drop(['id_x' , 'brand_id'] , axis=1)
df.rename(columns={'name' : 'brand' , 'id_y' : 'id'} , inplace=True)
df.head()
```

```python
df_2 = df.merge(os , left_on='os_id' , right_on='id')
df_2 = df_2.drop(['id_y' , 'os_id'] , axis=1)
df_2.rename(columns={'type' : 'os_type' , 'version' : 'os_version' , 'id_x' : 'id'} , inplace=True)
df_2
```

```python
df_network = phone_network.merge(network , left_on = 'network_id' , right_on = 'id')
df_network.drop(['network_id' , 'id'] , axis=1 , inplace=True)

df_network
```

```python
def check_2g(x):
    mask = df_network[df_network['phone_id'] == x.id]
    mask = mask[mask['name'] == '2G']
    
    try:
        return int(x.id) == int(mask.phone_id)
    except:
        return False
    
def check_3g(x):
    mask = df_network[df_network['phone_id'] == x.id]
    mask = mask[mask['name'] == '3G']
    
    try:
        return int(x.id) == int(mask.phone_id)
    except:
        return False
    
def check_4g(x):
    mask = df_network[df_network['phone_id'] == x.id]
    mask = mask[mask['name'] == '4G']
    try:
        return int(x.id) == int(mask.phone_id)
    except:
        return False
```

```python
df_3 = df_2.copy()

df_3['2G'] = df_3.apply(check_2g , axis=1)
df_3['3G'] = df_3.apply(check_3g , axis=1)
df_3['4G'] = df_3.apply(check_4g , axis=1)

df_3
```

```python
df_3.to_csv('data/cleanned_data.csv', index=False)
```
