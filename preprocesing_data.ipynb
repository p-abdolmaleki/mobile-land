```python
import pandas as pd
```

```python
df_new = pd.read_csv('datas.csv')
```

```python
df_new['memory'] = df_new['memory'].apply(eval)
```

```python
df_new.reset_index(inplace=True)
df_new.rename(columns={'index':'id'}, inplace=True)
```

```python
df_new['id'] = df_new['id'] + 1
```

```python
length = list()
width = list()
thinckness = list()
for row in df_new['dimensions'].str.split(' x '):
    length.append(row[0])
    width.append(row[1])
    thinckness.append(row[2].strip())
```

```python
df_new['length'] = length
df_new['width'] = width
df_new['thinckness'] = thinckness
```

```python
length_resolution = list()
width_resolution = list()
for row in df_new['resolution'].str.split(' x '):
    length_resolution.append(row[0])
    width_resolution.append(row[1].strip())
```

```python
df_new['length_resolution'] = length_resolution
df_new['width_resolution'] = width_resolution
```

```python
df_new.loc[df_new['type'] == 'iPadOS', 'type'] = 'iOS'
```

```python
df_new['type'].unique()
```

```python
os = list()
for row in df_new[['type', 'version']].values:
   os.append((row[0],row[1]))
```

```python
os = set(os)
```

```python
os_id = list()
for row in df_new[['type', 'version']].values:
    for i, val in enumerate(os):
        if val == ((row[0], row[1])):
            os_id.append(i + 1)
```

```python
df_new['os_id'] = os_id
```

```python
df_new[['os_id', 'type', 'version', ]]
```

```python
brand = list()
for row in df_new['brand']:
    brand.append(row)
```

```python
brand
```

```python
brand = set(brand)
```

```python
brand_id = list()
for row in df_new['brand']:
    for i, val in enumerate(brand):
        if val == row:
            brand_id.append(i + 1)
```

```python
brand_id
```

```python
df_new['brand_id'] = brand_id
```

```python
phone = df_new[['id', 'brand_id', 'name', 'year', 'price', 'battery', 'network', 'os_id', 'weight', 'sim', 'size', 'length_resolution', 'width_resolution', 'ppi_density', 'screen_to_body_ratio', 'length', 'width', 'thinckness']]
```

```python
phone.head()
```

```python
brand_with_id = list()
for row in df_new[['brand_id', 'brand']].values:
    brand_with_id.append((row[0], row[1]))
```

```python
brand_with_id = set(brand_with_id)
```

```python
brand = pd.DataFrame(brand_with_id)
brand = brand.rename(columns={0:'id', 1:'name'})
brand
```

```python
os_with_id = list()
for row in df_new[['os_id', 'type', 'version']].values:
    os_with_id.append((row[0], row[1], row[2]))
```

```python
os_with_id = set(os_with_id)
```

```python
os = pd.DataFrame(os_with_id)
os.rename(columns={0:'id', 1:'type', 2:'version'}, inplace=True)
os
```

```python
df_new['sensor'] = df_new['sensor'].apply(eval)
```

```python
sensor_list = list()
for row in df_new['sensor'].values:
    for i in row:
        sensor_list.append(i)
```

```python
sensor_set = set(sensor_list)
```

```python
sensor_with_id = list()
for row in df_new['sensor']:
    new_list = list()
    for val in row:
        for i, sens in enumerate(sensor_set):
            if val == sens:
                new_list.append((i + 1, val))
    sensor_with_id.append(new_list)
```

```python
sensor_with_id
```

```python
df_new['sensor_with_id'] = sensor_with_id
```

```python
df_new[['id', 'sensor_with_id']]
```

```python
phone_sensor = list()
for row in df_new[['id', 'sensor_with_id']].values:
    for val in row[1]:
        phone_sensor.append((row[0],val[0]))
```

```python
phone_sensor = pd.DataFrame(phone_sensor)
phone_sensor.rename(columns={0:'phone_id', 1:'sensor_id'}, inplace=True)
phone_sensor.head()
```

```python
sensor = list()
for row in sensor_with_id:
    for val in row:
        sensor.append(val)
```

```python
sensor = set(sensor)
```

```python
sensor = pd.DataFrame(sensor)
sensor.rename(columns={0:'id', 1:'name'}, inplace=True)
sensor.head()
```

```python
memories = list()
for row in df_new['memory']:
    for val in row:
        memories.append(val)
```

```python
memories = set(memories)
```

```python
mem_with_id = list()
for row in df_new['memory']:
    new_list = list()
    for val in row:
        for i, mem in enumerate(memories):
            if mem == val:
                new_list.append((i+1, mem))
    mem_with_id.append(new_list)
```

```python
for row in df_new['memory']:
    print(row)
```

```python
df_new['mem_with_id'] = mem_with_id
```

```python
phone_memory = list()
for row in df_new[['id', 'mem_with_id']].values:
    for val in row[1]:
        phone_memory.append((row[0], val[0]))
```

```python
phone_memory = pd.DataFrame(phone_memory)
phone_memory.rename(columns={0:'phone_id', 1:'memory_id'}, inplace=True)
```

```python
phone_memory
```

```python
memory = list()
for row in df_new['mem_with_id'].values:
    for val in row:
        memory.append((val[0], val[1][0], val[1][1]))
```

```python
memory = set(memory)
```

```python
memory = pd.DataFrame(memory)
memory.rename(columns={0:'id', 1:'main_memory', 2:'ram'}, inplace=True)
memory.head()
```

```python
phone.rename(columns={'name':'model', 'ppi_density':'ppi', 'screen_to_body_ratio':'ratio'}, inplace=True)
phone.head()
```

```python
phone.to_csv('db_datas/phone.csv', index=False)
os.to_csv('db_datas/os.csv', index=False)
brand.to_csv('db_datas/brand.csv', index=False)
phone_sensor.to_csv('db_datas/phone-sensor.csv', index=False)
sensor.to_csv('db_datas/sensor.csv', index=False)
phone_memory.to_csv('db_datas/phone-memory.csv', index=False)
memory.to_csv('db_datas/memory.csv', index=False)
df_new.to_csv('db_datas/preprocesing_datas.csv', index=False)
```
